## Simulator

```python
messages = {
    "reasoning_content": (
        "<think>\n"
        "Fibonacci sequence is defined as F(n) = F(n-1) + F(n-2) where F(0) = 0 and F(1) = 1.\n"
        "The recursive solution is elegant but inefficient due to repeated calculations.\n"
        "The iterative solution has O(n) time complexity and O(1) space complexity.\n"
        "</think>\n"
    ),
    "content": (
        "Here's how to calculate Fibonacci numbers:\n"
        "I should provide both recursive and iterative solutions\n"
        "<code>\n"
        "def fibonacci_recursive(n):\n"
        "    if n <= 1:\n"
        "        return n\n"
        "    return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)\n"
        "</code>\n"
        "The recursive solution has exponential time complexity.\n"
        "<code>\n"
        "def fibonacci_iterative(n):\n"
        "    a, b = 0, 1\n"
        "    for _ in range(n):\n"
        "        a, b = b, a + b\n"
        "    return a\n"
        "</code>\n"
        "The iterative solution is more efficient."
    ),
}
```

```python
import random 
import time 
from typing import Dict, Generator

class StreamingSimulator:
    def __init__(self, 
                 messages = Dict[str, str],
                 delay_range: tuple = (0.01, 0.05), 
                 chunk_size_range: tuple = (1, 10),):
        self.messages = messages
        self.delay_range = delay_range
        self.chunk_size_range = chunk_size_range
        
    def stream_response(self) -> Generator[Dict[str, str], None, None]:
        chunk_size = random.randint(*self.chunk_size_range)
        for key, message in self.messages.items():
            position = 0 
            while position < len(message):
                yield {key: message[position:position+chunk_size]}
                position += chunk_size
                time.sleep(random.uniform(*self.delay_range))
```

```python
stream_simulator = StreamingSimulator(messages)
accumulated = ""
for chunk in stream_simulator.stream_response():
    delta = chunk["content"] if "content" in chunk else chunk["reasoning_content"]
    print(delta, end="", flush=True)
    accumulated += delta
```
